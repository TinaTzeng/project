```python
from nba_api.stats.endpoints import playercareerstats, TeamDashboardByTeamPerformance, LeagueSeasonMatchups, PlayerGameLog, PlayerGameLogs
from nba_api.stats.endpoints import leaguegamefinder, PlayerDashboardByTeamPerformance, TeamInfoCommon, CommonPlayerInfo
from nba_api.stats.static import teams
from nba_api.stats.static import players
import pandas as pd
```

# 所有球隊資訊

```python
#get teams
all_teams = teams.get_teams()
print(type(all_teams))
len(all_teams)
#總共30支隊伍
```

```python
columns = [k for k in all_teams[0].keys()]
print(columns)
```

```python
rows = []
for team in all_teams:
    rows.append([v for v in team.values()])
#     print([v for v in team.values()])
# print(rows)
```

```python
# 列出所有隊伍資料
df = pd.DataFrame(data = rows, columns = columns)
# df.index = df.index + 1
# df.index.name = 'TeamNum'
df
```

```python
df.to_csv(r'C:\Users\ACER\Python專案\jupyter\nbaData\all_teams.csv', index = False)
```

# 取出某特定隊伍的比賽資訊

```python
# 取出特定隊伍的比賽資訊
# leaguegamefinder: find team games 
# team_id : df.iloc[0,0]
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=df.iloc[0,0], season_segment_nullable = '2019-20')
games = gamefinder.get_data_frames()
print(type(games))
print(len(games))
games = gamefinder.get_data_frames()[0]
games.head()
# games.columns
```

# 列出所有active球員

```python
#列出所有球員
all_players = players.get_players()
active_players = players.get_active_players()
print(type(active_players))
print('all players:',len(all_players))
print('active players:',len(active_players))
```

```python
columns_p = [k for k in all_players[0].keys()]
print(columns_p)
```

```python
# print([v for v in active_players[0].values()])
rows_p = []
for player in all_players:
#     print([v for v in player.values()])
    rows_p.append([v for v in player.values()])
# print(rows_p)
```

```python
# 取出所有球員資料
df_player = pd.DataFrame(data = rows_p, columns = columns_p)
df_player.head()
```

```python
df_player.to_csv(r'C:\Users\ACER\Python專案\jupyter\nbaData\all_players.csv', index = False)
```
